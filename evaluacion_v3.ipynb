```python
import pandas as pd

df_train = pd.read_csv('tsp/train_df.csv', index_col=0)
df_eval = pd.read_csv('tsp/eval_df.csv', index_col=0)
```

```python
df_eval
```

```python
## Deviacion estandar para costos promedio
eval_stdd = df_eval.groupby(["model_name", "fold"]).agg(['mean', 'std']).reset_index()
eval_stdd
```

```python
train_stdd = df_train.groupby(["model_name", "epoch"]).agg(
    {
        "tr_loss": ["mean", "std"],
        "tr_acc": ["mean", "std"],
        "val_loss": ["mean", "std"],
        "val_acc": ["mean", "std"]
    }
).reset_index()
train_stdd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the plot
metrics = ["tr_loss", "tr_acc", "val_loss", "val_acc"]
colors = sns.color_palette("tab10", len(train_stdd["model_name"].unique()))
models = train_stdd["model_name"].unique()

fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True)

for i, metric in enumerate(metrics):
    ax = axes[i // 2, i % 2]
    for model, color in zip(models, colors):
        model_data = train_stdd[train_stdd["model_name"] == model]
        mean_values = model_data[metric]['mean']
        std_values = model_data[metric]['std']
        
        # Line plot for the mean
        ax.plot(
            model_data["epoch"],
            mean_values,
            marker='o',
            label=model,
            color=color
        )
        
        # Shaded area for the standard deviation
        ax.fill_between(
            model_data["epoch"],
            mean_values - std_values,
            mean_values + std_values,
            color=color,
            alpha=0.3
        )
        
    ax.set_title(metric)
    ax.set_xlabel("epoch")
    ax.set_ylabel(metric)
    ax.legend(title="model_name")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

Se observa en todos los modelos una tendencia a converger tanto en Training Loss como Training Accuracy. 

En cuanto a las métricas de validación , hay una mejora continua a lo largo de todos los modelos.

MultilayerTransformerBlockModel pareciera tener la menor desviación estándar respecto a los otros modelos
