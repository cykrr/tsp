```python
import pandas as pd

df_train = pd.read_csv('tsp/train_df.csv', index_col=0)
df_eval = pd.read_csv('tsp/eval_df.csv', index_col=0)

df_base_train = pd.read_csv('base_train_v2.csv', index_col=0)
df_base_eval = pd.read_csv('base_eval_v2.csv', index_col=0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_base_train.groupby(['Epoch']).mean()
```

```python
df_train[df_train["model_name"] == "BaseModel"].groupby(['epoch', 'model_name']).mean().head().reset_index()
```

```python
metric_name_a = ["Training Loss", "Training Accuracy", "Validation Loss", "Validation Accuracy"]
metric_name_b = ["tr_loss", "tr_acc", "val_loss", "val_acc"]
dfs = [
    df_train[df_train["model_name"] == "BaseModel"].groupby(['epoch', 'model_name']).mean().head().reset_index(), # Previo
    df_base_train.groupby(['Epoch']).mean().reset_index() # Profe 
]
names = ["Kfold", "Profe"]
metrics = [metric_name_b, metric_name_a]
colors = sns.color_palette("tab10", len(metrics))
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True)

for j in range(len(metric_name_a)):
    ax = axes[j // 2, j % 2]
    for i in range(len(metrics)):
        model_data = dfs[i]
        mean_values = model_data[metrics[i][j]]
        ax.plot(
            model_data['epoch' if i == 0 else 'Epoch'],
            mean_values,
            marker='o',
            label=names[i],
            color=colors[i]
        )
    ax.set_title(metrics[1][j])
    ax.set_xlabel("Epoch")
    ax.set_ylabel(metrics[1][j])
    ax.legend(title="Modelo")
    ax.grid(True)
plt.tight_layout()
plt.show()
        
    
    
```

Se realizó un análisis  comparativo del entrenamiento en dos configuraciones, una con validación cruzada K-Fold y otra sin ella. Los resultados muestran diferencias significativas en la estabilidad y precisión de ambos enfoques.

### Fase de Entrenamiento
**Modelo con K-Fold:** Presenta una curva de pérdida (loss) más estable durante el entrenamiento, con un descenso progresivo y consistente. Esto indica que el modelo generaliza mejor sobre los datos de entrenamiento al aprovechar múltiples particiones para validar y ajustar los pesos en cada iteración.

**Modelo sin K-Fold:**
La pérdida es más irregular, con fluctuaciones notables a lo largo del proceso de entrenamiento. Si bien puede alcanzar valores bajos en algunos momentos, estas oscilaciones reflejan una menor robustez en la convergencia.

### Fase de Validación
**Modelo con K-Fold:**
Mantiene una reducción estable de la pérdida en el conjunto de validación, lo cual sugiere una mejor capacidad de generalización en múltiples subconjuntos de datos. Sin embargo, la precisión final alcanzada es ligeramente inferior en comparación con el modelo sin K-Fold, lo que podría implicar una compensación entre estabilidad y ajuste a los datos específicos de validación.


**Modelo sin K-Fold:**
A pesar de la irregularidad en el entrenamiento, este modelo logra alcanzar una mayor precisión al final de la validación. Esto puede interpretarse como un ajuste más específico a los datos del conjunto de validación, aunque a expensas de una menor estabilidad.
Conclusiones
El uso de K-Fold ofrece ventajas significativas en términos de estabilidad y generalización, lo que es ideal para problemas donde la robustez es clave. Por otro lado, el modelo sin K-Fold, aunque más susceptible a variaciones, puede adaptarse mejor a conjuntos de validación pequeños o específicos.
